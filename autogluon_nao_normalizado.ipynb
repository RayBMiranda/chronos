```python
import pandas as pd
from autogluon.timeseries import TimeSeriesPredictor

# 1. Carregar o conjunto de dados
# O arquivo deve conter colunas: 'timestamp', 'ph', 'temperature', 'conductivity', 'solids', 'turbidity'.
data = pd.read_csv('water_quality.csv')

# Verificar os primeiros dados
print(data.head())

# 2. Preparar o conjunto de dados
# Garantir que 'timestamp' esteja em formato de data
data['timestamp'] = pd.to_datetime(data['Data'])

# Ordenar por timestamp (recomendado para séries temporais)
data = data.sort_values('timestamp')

# Configurar o índice como o timestamp
#data = data.set_index('timestamp')

data.drop(columns=['Data'], inplace=True)


data["item_id"] = 1

# 3. Configuração do AutoGluon TimeSeries
predictor = TimeSeriesPredictor(
    target="Turbidity",  # Variável alvo
    prediction_length=10,  # Previsão de 10 unidades de tempo à frente
    #time_limit=300,  # Tempo de treinamento (em segundos)
    known_covariates_names=['pH', 'Alkalinity', 'RainAccumulation15D', 'RiverLevel'],  # Variáveis exógenas
)

# 4. Treinar o modelo
predictor.fit(data)

```
